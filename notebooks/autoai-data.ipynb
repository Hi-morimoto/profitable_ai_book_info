##  AutoAIデータ作成用

```python
# 共通事前処理

# 必要ライブラリのimport
import pandas as pd
# データフレーム表示用関数
from IPython.display import display
```

```python
# 公開データのダウンロードと解凍
!wget https://archive.ics.uci.edu/ml/\
machine-learning-databases/00222/bank.zip -O bank.zip | tail -n 1
!unzip -o bank.zip | tail -n 1

# bank-full.csvをデータフレームに取り込み
df_all = pd.read_csv('bank-full.csv', sep=';')

# 項目名を日本語に置き換える
columns = [
    '年齢', '職業', '婚姻', '学歴', '債務不履行',  '平均残高',  '住宅ローン',  '個人ローン',
    '連絡手段',  '最終通話日',  '最終通話月',  '最終通話秒数', '通話回数_販促中',
    '前回販促後_経過日数',  '通話回数_販促前',  '前回販促結果', '今回販促結果'
]
df_all.columns = columns
```

```python
display(df_all.head())
```

```python
# CSV export
fn = 'bank-autoai.csv'
df_all.to_csv(fn, index=False)
```
